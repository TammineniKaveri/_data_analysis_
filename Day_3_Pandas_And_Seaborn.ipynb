Linear Regression algorithm
* plotting the data set

Linear Fit
* lmplot()
* a linear line is rotated to a particular angle until the distance between individual points in a graph that are plotted and the tilted line is minimum in all cases

```python
sns.lmplot(x="size",y="tip",data=tips);
```

```python
sns.lmplot(data=fmri,x="timepoint",y="signal")
```
